```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set x axis to numerical value for month
x_axis_data = np.arange(1,13,1)
x_axis_data
```

```python
# Average weather temp
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
plt.plot(x_axis_data,points )
plt.show()
```

```python
points
```

```python
#convert to celsius without list comprehension
points_c  = []
for temp in points:
    temp_c = (temp -32) * 0.56
    points_c.append(temp_c)
#c = F-32 *.56
points_c
```

```python
    #same with a list comprehension
    points_c2 = [(temp -20) * 0.56 for temp in points]
    points_c2
```

```python
c_handle, = plt.plot(x_axis_data,points_c,marker='o',color="red",label="celsius" )
t_handle, = plt.plot(x_axis_data,points, marker='*',color="blue" ,label="faren...")

plt.xlabel("Month Number")
plt.ylabel("average temp")

plt.hlines(0,0,12, alpha=.25)

plt.legend()
plt.show()
plt.savefig("my_pic_1.png")
```
