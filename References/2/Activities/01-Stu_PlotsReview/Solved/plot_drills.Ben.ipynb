```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA SET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
plt.bar(gyms, members)
plt.xlabel("Gym Name")
plt.ylabel("Number of Members")
plt.title("Gyms in New York")
```

```python
# DATA SET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.plot(x_axis,sin, marker="o", linewidth=0, color="red")
plt.title("is it a sin or a sign")
plt.xlabel("numbers")
plt.ylabel("sin(x)")
```

```python
# DATA SET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.pie(members, labels=gyms, startangle=90, colors=colors,
        explode=explode,shadow=True, autopct="%1.1f%%")
plt.title("Gyms in New York")
```

```python
# DATA SET 4
x_axis = np.arange(0, 10, 0.1)
times = []
sizes = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
    sizes.append(np.random.randint(1,550))
```

```python
plt.scatter(x_axis, times,color="red",s=sizes)
plt.title("Fake Time Data")
plt.xlabel("time something happened")
plt.ylabel("Number of things at that time ")
```
